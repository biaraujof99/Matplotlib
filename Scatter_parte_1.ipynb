```python
# Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X
X = np.linspace(0,5,11)
```

```python
# Criando funções Y(X) e Z(X)
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
# Gráfico de Y em função de X
plt.scatter(X,Y)
plt.show()
```

```python
# Inserindo nomes nos eixos
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
#  Inserindo título no gráfico
plt.scatter(X,Y)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.show()
```

```python
# Dois gráficos na mesma imagem
plt.scatter(X,Y)
plt.scatter(X,Z)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Alterando cores
plt.scatter(X,Y,color='blue')
plt.scatter(X,Z,color='orange')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
# Inserindo legenda
plt.scatter(X,Y,color='black',label='Y(X)')
plt.scatter(X,Z,color='red',label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
